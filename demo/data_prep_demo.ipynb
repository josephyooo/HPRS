```python
import matplotlib.pyplot as plt
from torchvision.io import read_image
from torchvision.utils import save_image
from torchvision.models import GoogLeNet_Weights
import torchvision.transforms.v2 as v2
```

```python
sample_image = read_image("../Image_Data/Alopecia-pictures/alopecia-areata-14.jpg")

plt.imshow(sample_image.permute(1, 2, 0))

save_image(sample_image/255, "sample_image.jpg")
```

```python
transformed_image = GoogLeNet_Weights.DEFAULT.transforms()(sample_image)

plt.imshow(transformed_image.permute(1, 2, 0))

save_image(transformed_image, "transformed_image.jpg")
```

```python
transformed_augmented_image = v2.RandAugment()(transformed_image)

plt.imshow(transformed_augmented_image.permute(1, 2, 0))

save_image(transformed_augmented_image, "transformed_augmented_image.jpg")
```
